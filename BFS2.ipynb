```python
# Simple BFS
# Can try with deque from collections
# Adapt to grids
%matplotlib inline
import networkx as nx
```

```python
def BFS(graph, start_node):
    visited = set()
    node_list = [start_node]

    while len(node_list) > 0:
        node = node_list.pop(0)
        
        print(node)
        visited.add(node)
        for next in graph.edges([node]):
            if next[1] not in visited:
                node_list.append(next[1])
```

```python
G = nx.Graph({
    '1': ['2', '3', '4'],
    '2': ['5'],
    '3': ['6'],
    '4': ['7'],
    '5': ['8', '9'],
    '6': ['10'],
    '7': [],
    '8': [],
    '9': [],
    '10': []
})

# BFS(G, '5')

G.edges(['5'])
nx.draw(G, with_labels=True)
```
